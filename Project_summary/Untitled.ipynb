```python
import os
import numpy as np
from matplotlib import pyplot as plt
import pytraj as pt
import pandas as pd
import seaborn as sns
```

```python
APAP_POX_traj1=pt.iterload('APAP_1C_2P_0-300ns_strip.nc','APAP_strip.parm7')
APAP_POX_traj2=pt.iterload('APAP_1C_2P_rep2.nc','APAP_strip.parm7') 
APAP_POX_traj3=pt.iterload('APAP_1C_2P_rep3.nc','APAP_strip.parm7')
AA_POX_traj1=pt.iterload('4-AA_1C_2P_0-300ns_strip.nc','4-AA_strip.parm7')
AA_POX_traj2=pt.iterload('4-AA_1C_2P_rep2.nc','4-AA_strip.parm7')
AM404_POX_traj1=pt.iterload('AM404_1C_2P_0-300ns_strip.nc','AM404_strip.parm7')
AM404_POX_traj2=pt.iterload('AM404_1C_2P_rep2.nc','AM404_strip.parm7')
```

```python
dist1 = pt.distance(APAP_POX_traj1, ':1 :555')
dist2 = pt.distance(APAP_POX_traj1, ':2 :908')
dist3 = pt.distance(APAP_POX_traj2, ':1 :555')
dist4 = pt.distance(APAP_POX_traj2, ':2 :908')
dist5 = pt.distance(APAP_POX_traj3, ':1 :555')
dist6 = pt.distance(APAP_POX_traj3, ':2 :908')
```

```python
dist7 = pt.distance(AA_POX_traj1, ':2 :555')
dist8 = pt.distance(AA_POX_traj1, ':1 :908')
dist9 = pt.distance(AA_POX_traj2, ':2 :555')
dist10 = pt.distance(AA_POX_traj2, ':1 :908')
```

```python
dist11 = pt.distance(AM404_POX_traj1, ':2 :555')
dist12 = pt.distance(AM404_POX_traj1, ':1 :908')
dist13 = pt.distance(AM404_POX_traj2, ':2 :555')
dist14 = pt.distance(AM404_POX_traj2, ':1 :908')
```

```python
plt.figure(figsize = (20,8))
plt.plot(np.mean(dist1.reshape(300,20), axis=1),label="APAP at POX site rep1")
plt.plot(np.mean(dist3.reshape(300,20), axis=1),label="APAP at POX site rep2")
#plt.plot(np.mean(dist5.reshape(300,20), axis=1),label="APAP at POX site rep3")
plt.plot(np.mean(dist7.reshape(300,20), axis=1),label="4-AA at POX site rep1")
plt.plot(np.mean(dist9.reshape(300,20), axis=1),label="4-AA at POX site rep2")
plt.plot(np.mean(dist11.reshape(300,20), axis=1),label="AM404 at POX site rep1")
plt.plot(np.mean(dist13.reshape(300,20), axis=1),label="AM404 at POX site rep2")


plt.xlabel('Time /ns',fontsize=20)
plt.ylabel('Distance /',fontsize=20)
plt.title(' Distance between ligands and heme unit vs Time',fontsize=32)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5),prop={'size': 20})
plt.xticks(fontsize=16)
plt.yticks(range(0,101,10),fontsize=16)
plt.savefig("dist_POX.png",dpi=400,bbox_inches='tight')
```

```python
plt.figure(figsize = (20,8))
plt.plot(np.mean(dist6.reshape(300,20), axis=1),label="APAP at COX site rep1")
plt.plot(np.mean(dist4.reshape(300,20), axis=1),label="APAP at COX site rep2")
#plt.plot(np.mean(dist5.reshape(300,20), axis=1),label="APAP at POX site rep3")
plt.plot(np.mean(dist8.reshape(300,20), axis=1),label="4-AA at COX site rep1")
plt.plot(np.mean(dist10.reshape(300,20), axis=1),label="4-AA at COX site rep2")
plt.plot(np.mean(dist12.reshape(300,20), axis=1),label="AM404 at COX site rep1")
plt.plot(np.mean(dist14.reshape(300,20), axis=1),label="AM404 at COX site rep2")


plt.xlabel('Time /ns',fontsize=20)
plt.ylabel('Distance /',fontsize=20)
plt.title(' Distance between ligands and Tyr385 residue vs Time',fontsize=32)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5),prop={'size': 20})
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.savefig("dist_COX.png",dpi=400,bbox_inches='tight')
```
